```python
prj = xnat.select.project('CCIR_00754')
print(prj.id() + ' ' + prj.label())
```

```python
sbj = prj.subject('HYGLY48')
print(sbj.id() + ' ' + sbj.label())
for isbj in prj.subjects('*'):
    print(isbj.id() + ' ' + isbj.label())
```

```python
ses = sbj.experiment('CNDA_E249152')
print(str(ses)[20:])
print(sbj.experiments('CNDA_E249152').get()[0])
for ie in sbj.experiments('*'):
    print('--')
    print(str(ie)[20:])
    print(ie.id())
    print(ie.label())
```

```python
scn = ses.scan('2')
print(str(scn)[13:])
print(ses.scans('2').get()[0])
for isc in ses.scans('*'):
    print('--')
    print(str(isc)[13:])
    print(isc.id())
    print(isc.label())
```
